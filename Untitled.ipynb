```python
import pandas as pd
import pandas_datareader.data as web
import datetime

# Define the time period
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2023, 8, 30)

# Download Effective Federal Funds Rate from FRED
data = web.DataReader('DFF', 'fred', start, end)
data = data.dropna()

# Plot the data
data.plot(title='Effective Federal Funds Rate')
```

```python
import numpy as np
from scipy.optimize import minimize

# Calculate daily returns
data['returns'] = data['DFF'].pct_change().dropna()

# Define the negative log-likelihood function for Vasicek model
def vasicek_neg_log_likelihood(params, data):
    alpha, beta, sigma = params
    dt = 1/252  # Daily data
    
    r_t = data['DFF'].values
    dr_t = np.diff(r_t)
    r_t = r_t[:-1]
    
    # Calculate the log-likelihood
    n = len(dr_t)
    term1 = -0.5 * n * np.log(2 * np.pi)
    term2 = -n * np.log(sigma)
    term3 = -np.sum((dr_t - alpha * (beta - r_t) * dt) ** 2) / (2 * sigma ** 2)
    
    log_likelihood = term1 + term2 + term3
    return -log_likelihood

# Initial parameter guesses
initial_params = [0.1, data['DFF'].mean(), data['returns'].std()]

# Bounds and constraints for the parameters
bounds = [(0.01, 1), (0.01, 5), (0.01, 1)]

# Estimate the parameters using SLSQP
result = minimize(vasicek_neg_log_likelihood, initial_params, args=(data), method='SLSQP', bounds=bounds)
alpha, beta, sigma = result.x

print(f'Estimated Parameters: alpha={alpha}, beta={beta}, sigma={sigma}')

# Check if the optimization was successful
if result.success:
    print('Optimization was successful.')
else:
    print('Optimization failed.')
    print(result.message)
```

```python
import matplotlib.pyplot as plt

def vasicek_simulation(alpha, beta, sigma, r0, T, dt=1/252, n=1000):
    t = np.arange(0, T, dt)
    r = np.zeros_like(t)
    r[0] = r0
    
    for i in range(1, len(t)):
        dr = alpha * (beta - r[i-1]) * dt + sigma * np.sqrt(dt) * np.random.normal()
        r[i] = r[i-1] + dr
    
    return r

# Define the time step
dt = 1/252  # Daily data

# Simulate the interest rate paths
T = 1  # 1 year
n_simulations = 10
simulated_rates = np.zeros((int(T/dt), n_simulations))

for i in range(n_simulations):
    simulated_rates[:, i] = vasicek_simulation(alpha, beta, sigma, data['DFF'].iloc[-1], T, dt)

# Plot the simulated paths
plt.plot(simulated_rates)
plt.title('Simulated Vasicek Interest Rate Paths')
plt.xlabel('Days')
plt.ylabel('Interest Rate')
plt.show()
```

```python
import pandas as pd
import pandas_datareader.data as web
import datetime

# Define the time period
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2023, 9, 30)

# Download historical yield curve data from FRED (example using 1, 2, 5, 10, 20, and 30-year Treasury yields)
yields = web.DataReader(['DGS1', 'DGS2', 'DGS5', 'DGS10', 'DGS20', 'DGS30'], 'fred', start, end)
yields = yields.dropna()

# Plot the data
yields.plot(title='Historical Yield Curves')
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the Nelson-Siegel model
def nelson_siegel(t, beta0, beta1, beta2, lambda_):
    return beta0 + beta1 * (1 - np.exp(-lambda_ * t)) / (lambda_ * t) + beta2 * ((1 - np.exp(-lambda_ * t)) / (lambda_ * t) - np.exp(-lambda_ * t))

# Define the objective function for calibration
def ns_objective(params, maturities, yields):
    beta0, beta1, beta2, lambda_ = params
    fitted_yields = nelson_siegel(maturities, beta0, beta1, beta2, lambda_)
    return np.sum((yields - fitted_yields) ** 2)

# Example maturities (1, 2, 5, 10, 20, 30 years)
maturities = np.array([1, 2, 5, 10, 20, 30])

# Use the most recent yield data for calibration
yields_recent = yields.iloc[-1].values

# Initial parameter guesses
initial_params = [0.03, -0.02, 0.02, 0.1]

# Perform calibration using SLSQP
result = minimize(ns_objective, initial_params, args=(maturities, yields_recent), method='SLSQP')
beta0, beta1, beta2, lambda_ = result.x

print(f'Calibrated Parameters: beta0={beta0}, beta1={beta1}, beta2={beta2}, lambda={lambda_}')

# Plot the fitted yield curve
fitted_yields = nelson_siegel(maturities, beta0, beta1, beta2, lambda_)
plt.plot(maturities, yields_recent, 'o', label='Observed Yields')
plt.plot(maturities, fitted_yields, '-', label='Fitted Nelson-Siegel Curve')
plt.title('Calibrated Nelson-Siegel Yield Curve')
plt.xlabel('Maturity (Years)')
plt.ylabel('Yield')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize

def vasicek_yield_curve(alpha, beta, sigma, maturities, r0):
    B = (1 - np.exp(-alpha * maturities)) / alpha
    A = np.exp((beta - sigma**2 / (2 * alpha**2)) * (B - maturities) - (sigma**2 / (4 * alpha)) * B**2)
    P = A * np.exp(-B * r0)
    yields = -np.log(P) / maturities
    return yields

def objective(params, maturities, observed_yields, r0):
    alpha, beta, sigma = params
    model_yields = vasicek_yield_curve(alpha, beta, sigma, maturities, r0)
    return np.sum((observed_yields - model_yields) ** 2)

# Define maturities and observed yields
maturities = np.array([1, 2, 5, 10, 20, 30])
observed_yields = yields.loc['2023-08-29'].values   # Convert to decimal

# Initial short-term rate
r0 = yields.loc['2023-08-29'].values[0]  # Convert to decimal

# Initial parameter guesses
initial_params = [0.1, observed_yields.mean(), observed_yields.std()]

# Bounds for the parameters
bounds = [(0.01, 1), (0.01, 10), (0.01, 1)]

# Optimize the parameters
result = minimize(objective, initial_params, args=(maturities, observed_yields, r0), method='L-BFGS-B', bounds=bounds)
alpha, beta, sigma = result.x

print(f'Calibrated Parameters: alpha={alpha}, beta={beta}, sigma={sigma}')
```

```python
# Generate the yield curve using the calibrated parameters
calibrated_yields = vasicek_yield_curve(alpha, beta, sigma, maturities, r0)

# Plot the observed yields vs. the calibrated model yields
import matplotlib.pyplot as plt

plt.plot(maturities, observed_yields, 'o', label='Observed Yields (29th Sep 2023)')
plt.plot(maturities, calibrated_yields, '-', label='Calibrated Vasicek Yield Curve')
plt.title('Observed vs. Calibrated Yield Curve')
plt.xlabel('Maturity (Years)')
plt.ylabel('Yield')
plt.legend()
plt.show()
```
