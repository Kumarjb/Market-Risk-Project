```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Download historical data for the federal funds rate from FRED
start_date = '2001-01-01'
end_date = '2023-08-31'
data = web.DataReader('DFF', 'fred', start_date, end_date)

# Convert the data to a DataFrame and rename the column
interest_rates = data.rename(columns={'DFF': 'Rate'})

# Save the data to a CSV file
interest_rates.to_csv("federal_funds_rates.csv")

print("Data downloaded and saved successfully.")

# Calculate the daily changes in interest rates
interest_rates = interest_rates.dropna()  # Ensure no NaNs
delta_rates = interest_rates.diff().dropna()

# Define the Vasicek log-likelihood function
def vasicek_log_likelihood(params, rates, delta_rates):
    a, b, sigma = params
    dt = 1 / 252  # Assuming daily data with 252 trading days in a year
    r = rates[:-1]
    dr = delta_rates

    # Mean and variance of the normal distribution
    mean = a * (b - r) * dt
    variance = sigma ** 2 * dt

    # Log-likelihood
    log_likelihood = -np.sum(np.log(variance) + (dr - mean) ** 2 / variance)

    return -log_likelihood

# Initial parameter guesses
initial_params = [0.1, interest_rates['Rate'].mean(), interest_rates['Rate'].std()]

# Optimize the parameters
result = minimize(vasicek_log_likelihood, initial_params, args=(interest_rates['Rate'].values, delta_rates['Rate'].values))

# Extract the optimized parameters
a, b, sigma = result.x
print(f"Optimized parameters: a={a}, b={b}, sigma={sigma}")

# Function to calculate the theoretical yield curve using the Vasicek model
def vasicek_yield_curve(a, b, sigma, r0, maturities):
    dt = 1 / 252
    yields = []
    for T in maturities:
        B = (1 - np.exp(-a * T)) / a
        A = np.exp((b - sigma**2 / (2 * a**2)) * (B - T) - (sigma**2 * B**2) / (4 * a))
        yield_rate = (np.log(A) - np.log(r0)) / T + r0 * B / T
        yields.append(yield_rate)
    return np.array(yields)

# Define the maturities (in years)
maturities = np.array([1/12, 3/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30])  # 1 month, 3 months, 6 months, 1 year, etc.

# Calculate the theoretical yield curve
r0 = interest_rates['Rate'].iloc[-1]
theoretical_yields = vasicek_yield_curve(a, b, sigma, r0, maturities)

# Download real-time yield curve data from FRED (as of August end)
yield_data = web.DataReader(['DGS1MO', 'DGS3MO', 'DGS6MO', 'DGS1', 'DGS2', 'DGS3', 'DGS5', 'DGS7', 'DGS10', 'DGS20', 'DGS30'],
                            'fred', '2023-08-31', '2023-08-31')

# Prepare the observed yield curve
observed_yields = yield_data.iloc[0].values

# Plot the observed and theoretical yield curves
plt.figure(figsize=(10, 6))
plt.plot(maturities, theoretical_yields, label="Theoretical Yield Curve (Vasicek)", marker='o')
plt.plot(maturities, observed_yields, label="Observed Yield Curve", marker='x')
plt.title("Observed vs Theoretical Yield Curve")
plt.xlabel("Maturity (years)")
plt.ylabel("Yield (%)")
plt.legend()
plt.show()
```

```python
# Function to calculate the SSE between the observed and theoretical yield curves
def sse_vasicek(params, observed_yields, maturities, r0):
    a, b, sigma = params
    theoretical_yields = vasicek_yield_curve(a, b, sigma, r0, maturities)
    sse = np.sum((observed_yields - theoretical_yields) ** 2)
    return sse

# Recalibrate the Vasicek model parameters to match the observed yield curve
def recalibrate_vasicek(observed_yields, maturities, r0):
    # Initial parameter guesses
    initial_params = [0.1, 0.1, 0.1]
    
    # Parameter bounds
    bounds = [(0.0001, 1.0),  # bounds for 'a'
              (0.0001, 1.0),  # bounds for 'b'
              (0.0001, 0.5)]  # bounds for 'sigma'
    
    # Optimize the parameters to minimize the SSE
    result = minimize(sse_vasicek, initial_params, args=(observed_yields, maturities, r0), method='SLSQP', bounds=bounds)
    
    # Extract the optimized parameters
    a, b, sigma = result.x
    return a, b, sigma

# Define the maturities (in years)
maturities = np.array([1/12, 3/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30])  # 1 month, 3 months, 6 months, 1 year, etc.

# Download real-time yield curve data from FRED (as of August end)
yield_data = web.DataReader(['DGS1MO', 'DGS3MO', 'DGS6MO', 'DGS1', 'DGS2', 'DGS3', 'DGS5', 'DGS7', 'DGS10', 'DGS20', 'DGS30'],
                            'fred', '2023-08-31', '2023-08-31')

# Prepare the observed yield curve
observed_yields = yield_data.iloc[0].values

# Initial short-term rate
r0 = interest_rates['Rate'].iloc[-1]

# Recalibrate the Vasicek model parameters
a_recalibrated, b_recalibrated, sigma_recalibrated = recalibrate_vasicek(observed_yields, maturities, r0)

# Calculate the theoretical yield curve with recalibrated parameters
theoretical_yields_recalibrated = vasicek_yield_curve(a_recalibrated, b_recalibrated, sigma_recalibrated, r0, maturities)

# Plot the observed and recalibrated theoretical yield curves
plt.figure(figsize=(10, 6))
plt.plot(maturities, theoretical_yields_recalibrated, label="Recalibrated Theoretical Yield Curve (Vasicek)", marker='o')
plt.plot(maturities, observed_yields, label="Observed Yield Curve", marker='x')
plt.title("Observed vs Recalibrated Theoretical Yield Curve")
plt.xlabel("Maturity (years)")
plt.ylabel("Yield (%)")
plt.legend()
plt.show()

# Print the recalibrated parameters
print(f"Recalibrated parameters: a={a_recalibrated}, b={b_recalibrated}, sigma={sigma_recalibrated}")
```

```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.interpolate import interp1d

# Function to calculate the SSE between the observed and theoretical yield curves
def sse_vasicek(params, observed_yields, maturities, r0):
    a, b, sigma = params
    theoretical_yields = vasicek_yield_curve(a, b, sigma, r0, maturities)
    sse = np.sum((observed_yields - theoretical_yields) ** 2)
    return sse

# Recalibrate the Vasicek model parameters to match the observed yield curve
def recalibrate_vasicek(observed_yields, maturities, r0):
    # Initial parameter guesses
    initial_params = [0.1, 0.1, 0.1]
    
    # Parameter bounds
    bounds = [(0.0001, 1.0),  # bounds for 'a'
              (0.0001, 1.0),  # bounds for 'b'
              (0.0001, 0.5)]  # bounds for 'sigma'
    
    # Optimize the parameters to minimize the SSE
    result = minimize(sse_vasicek, initial_params, args=(observed_yields, maturities, r0), method='SLSQP', bounds=bounds)
    
    # Extract the optimized parameters
    a, b, sigma = result.x
    return a, b, sigma

# Function to calculate the theoretical yield curve using the Vasicek model
def vasicek_yield_curve(a, b, sigma, r0, maturities):
    yields = []
    for T in maturities:
        B = (1 - np.exp(-a * T)) / a
        A = np.exp((b - sigma**2 / (2 * a**2)) * (B - T) - (sigma**2 * B**2) / (4 * a))
        yield_rate = (-np.log(A) + (B * r0)) / T
        yields.append(yield_rate)
    return np.array(yields)

# Define the maturities (in years)
maturities = np.array([1/12, 3/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30])  # 1 month, 3 months, 6 months, 1 year, etc.

# Download real-time yield curve data from FRED (as of August end)
yield_data = web.DataReader(['DGS1MO', 'DGS3MO', 'DGS6MO', 'DGS1', 'DGS2', 'DGS3', 'DGS5', 'DGS7', 'DGS10', 'DGS20', 'DGS30'],
                            'fred', '2023-08-31', '2023-08-31')

# Prepare the observed yield curve
observed_yields = yield_data.iloc[0].values

# Initial short-term rate
r0 = interest_rates['Rate'].iloc[-1]

# Recalibrate the Vasicek model parameters
a_recalibrated, b_recalibrated, sigma_recalibrated = recalibrate_vasicek(observed_yields, maturities, r0)

# Calculate the theoretical yield curve with recalibrated parameters
theoretical_yields_recalibrated = vasicek_yield_curve(a_recalibrated, b_recalibrated, sigma_recalibrated, r0, maturities)

# Interpolate to create a smooth curve for recalibrated theoretical yields
interp_func = interp1d(maturities, theoretical_yields_recalibrated, kind='cubic')
smooth_maturities = np.linspace(maturities.min(), maturities.max(), 500)
smooth_theoretical_yields = interp_func(smooth_maturities)

# Plot the observed and recalibrated theoretical yield curves
plt.figure(figsize=(10, 6))
plt.plot(smooth_maturities, smooth_theoretical_yields, label="Recalibrated Theoretical Yield Curve (Vasicek)", linestyle='-', color='blue')
plt.plot(maturities, observed_yields, label="Observed Yield Curve", marker='x', linestyle='None', color='red')
plt.title("Observed vs Recalibrated Theoretical Yield Curve")
plt.xlabel("Maturity (years)")
plt.ylabel("Yield (%)")
plt.legend()
plt.show()

# Print the recalibrated parameters
print(f"Recalibrated parameters: a={a_recalibrated}, b={b_recalibrated}, sigma={sigma_recalibrated}")
```
